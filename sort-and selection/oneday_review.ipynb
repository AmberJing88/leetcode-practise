```python
# 279 perfect squares with lagrange's four-squares therom
def numSquares(n):
    while True:
        d, m = divmod(n, 4)
        if m:
            break
        else:
            n = d
    if n % 8 == 7:
        return 4
    elif (n**0.5).is_integer():
        return 1
    else:
        squares = [i**2 for i in range(1, (n**0.5)+1)]
        i,j = 0, len(squares)-1
        while i <=j:
            s = squares[i]+squares[j]
            if s<n:
                i+=1
            elif s>n:
                j-= 1
            else:
                return 2
        return 3
```

```python
# bfs method
def numSquares(n):
    if n < 2:
        return n
    lst = []
    i = 1
    while i*i <=n:
        lst.append(i*i)
        i += 1
    cnt = 0
    tocheck = {n}
    while tocheck:
        cnt += 1
        temp = set()
        for x in tocheck:
            for y in lst:
                if x ==y:
                    return cnt
                if x<y:
                    break
                temp.add(x-y)
        tocheck = temp
    return cnt
```

```python
#785 bipartite graph
def isBipartite(graph):
    colors = [0]*len(graph)
    def valid(graph, color, curr, colors):
        if colors[curr] !=0:
            return colors[curr] == color
        colors[curr] = color
        for i in graph[curr]:
            if not valid(graph, -1*color, i, colors):
                return False
        return True
    for i in range(len(graph)):
        if colors[i] == 0 and not valid(graph, 1, 0, colors):
            return False
    return True
```

```python
def isBipartite(graph):
    colors = [0]* len(graph)
    for i in range(len(graph)):
        if colors[i] !=0:
            continue
        queue, colors[i] = [i], 1
        while queue:
            t = queue.pop(0)
            for a in graph[t]:
                if colors[a] == colors[t]:
                    return False
                if colors[a] ==0:
                    colors[a] = -1 * colors[t]
                    queue.append(a)
    return True
```

```python
#bfs
def canfinish(num, edges):
    graph = {i:set() for i in range(num)}
    indeg = {i:0 for i in range(num)}
    for x, y in edges:
        graph[x] |= {y}
        indeg[x] += 1
    queue = [ i for i in range(num) if indeg[x]==0]
    visit = set(graph)
    while queue:
        node = queue.pop(0)
        for neigh in graph[node]:
            if neigh in visit:
                return False
            indeg[neigh] -= 1
            if not indeg[neigh]:
                visit.add(neigh)
                queue += [neigh]
    return len(visit)==num
    
```

```python
#210 
def findOrder(n, edges):
    graph = [[] for _ in range(n)]
    visit = [0 for _ in range(n)]
    res = []
    def dfs(node):
        if visit[node] == -1:
            return False
        if visit[node] == 1:
            return True
        visit[node] = -1
        for neigh in graph[node]:
            if not dfs(neigh):
                return False
        visit[node] =1
        res.append(node)
        return True
    for x, y in edges:
        graph[x].append(y)
    for i in range(n):
        if not dfs(i):
            return []
    return res
```

```python
#261 valid tree
def validtree(n, edges):
    parents = [[] for _ in range(n)]
    
    def find(x):
        if parents[x] ==0:
            return x
        parents[x] = find(parents[x])
        return parents[x]
    for a,b in edges:
        roota, rootb = find(a), find(b)
        if roota == rootb:
            return False
        parents[roota] = rootb
    return len(edges) == n-1
```
